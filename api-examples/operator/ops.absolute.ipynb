mindspore.ops.absolute(input) -〉 Tensor
- 输入：必须是mindspore的tensor类型。
- 返回：tensor类型

```python
import numpy as np
import mindspore as ms
import torch
x = np.array([-1.2,1.1])

y1 = ms.ops.absolute(ms.tensor(x))
y2 = torch.absolute(torch.tensor(x))

print ('mindspore output -> ',y1)
print ('torch     output -> ',y2)
```

ms和jax没有显示dtype以及类型

下面是计算复数的例子

```python
x1 = np.array([-1.2+1.1j])

y1 = ms.ops.absolute(ms.tensor(x1))
y2 = torch.absolute(torch.tensor(x1))

print ('mindspore output -> ',y1)
print ('torch     output -> ',y2)
```
